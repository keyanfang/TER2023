```python
import scispacy
import spacy
from spacy import displacy
from spacy.lang.fr import French
nlp = spacy.load("en_ner_bc5cdr_md")  # load existing spaCy model
```

```python
table= {'Entity':[], 'Class':[]}

for paragraph in document.paragraphs:
  print('================')
  print(paragraph.text)
  doc = nlp(paragraph.text)
  ent_bc = {}
  for x in doc.ents:
      ent_bc[x.text] = x.label_
  for key in ent_bc:
      table['Entity'].append(key)
      table['Class'].append(ent_bc[key])
```

```python
import pandas as pd

trans_df = pd.DataFrame(table)
trans_df.to_csv ('/content/drive/My Drive/resources/data/train1_spacy.csv', index=False)
```

```python
import csv
all_drug=[]
with open('/content/drive/My Drive/resources/data/train1_new_drug.csv', mode='r') as data:
  for line in csv.reader(data):
    #  print(line)
    drugs=line[2]
    drugs=drugs.replace("[","")
    drugs=drugs.replace("]","")
    drugs=drugs.replace("'","")
    drugs=drugs.replace("/ ","")
    drugs=drugs.split(", ")

    for drug in drugs:
     all_drug.append(drug)
     all_drug.append(drug.title())
     all_drug.append(drug.upper())
     all_drug.append(drug.lower())



all_drug
```

```python
from spacy.pipeline import EntityRuler

term_drug = dict(
    drugs=all_drug,
)
# term_drug

nlp_drug = spacy.blank("fr") # create blank Language class

# Sentencizer component, needed for negation detection
nlp_drug.add_pipe("eds.sentences")
# Matcher component
nlp_drug.add_pipe("eds.matcher", config=dict(terms=term_drug))
# Negation detection
nlp_drug.add_pipe("eds.negation")
# # Entity extraction pipelines
# nlp.add_pipe("eds.drugs", config=dict(term_matcher="simstring"))
```

```python
para_d=[]
drug_detect=[]

for paragraph in document.paragraphs:
  print('================')
  print(paragraph.text)

  doc = nlp_drug(paragraph.text)
  print(doc.ents)
  # displacy.render(doc, style="ent",jupyter=True)
  if doc.ents:
    para_d.append(paragraph.text)
    s=str(doc.ents)

    s=s.replace("(","")
    s=s.replace(")","")
    s=s.replace(",","")

    drug_detect.append(s)
    
print(drug_detect)
```

```python
all_reaction=[]
with open('/content/drive/My Drive/resources/data/train1_reaction.csv', mode='r') as data:
  for line in csv.reader(data):
    #  print(line)
    reactions=line[2]
    reactions=reactions.replace("[","")
    reactions=reactions.replace("]","")
    reactions=reactions.replace("'","")
    reactions=reactions.replace("/ ","")
    reactions=reactions.split(", ")
    # print(reactions)
    for reaction in reactions:
      all_reaction.append(reaction)
      all_reaction.append(reaction.title())
      all_reaction.append(reaction.upper())
      all_reaction.append(reaction.lower())


print(all_reaction)
```
